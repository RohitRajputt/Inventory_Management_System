```python
# Open the json file in read mode and convert into dict for further operation

fd = open("Inventory.json", 'r')
txt = fd.read()
fd.close()
txt1 = json.loads(txt)
```

```python
# Updation function for INVENTORY.json

def update():
    je = json.dumps(txt1)
    fe = open("Inventory.json", "w")
    fe.write(je + '\n')
    fe.close()
    
```

```python
# IN this func all the work is for CUSTOMER and after BILLING part updation of Inventory takes place
# And SALES.JSON file is also created
def customer():

# Asking user for selecting one catogory    
    
    print("****CATEGORIES**** \n1: Shiirts \n2: T-Shirts \n3: Casual Shoes \n4: Formal Shoes \n5: Watch \n")

    def puch(cat):
        if cat.lower() == "shirt" or cat.lower() == "shirts" or cat =='1':
            print("*** CHECK ORDER ID '101' TO '110' AND SELECT THE PRODUCT \n\n", txt1)

        elif cat.lower() == "t-shirt" or cat.lower == "t-shirts" or cat == '2':
            print("*** CHECK ORDER ID '111' TO '120' \n\n", txt1)

        elif cat.lower == "casual shoe" or cat.lower() == "casual shoes" or cat == '3':
            print("*** CHECK ORDER ID '121' TO '130' \n\n", txt1)

        elif cat == "sports shoes" or cat.lower() == "sports shoe" or cat == '4':
            print("*** CHECK ORDER ID '131' TO '140' \n\n", txt1)

        elif cat.lower == "watch" or cat.lower() == "watches" or cat == '5':
            print("*** CHECK ORDER ID '141' TO '150' \n\n", txt1)

        else:
            print("**** INVALID SELECTION****")
            print("*** ENTER VALID SELECTION***")

    cat = input("Emter your Category...")
    print('\n')
    puch(cat)


    ui = input("\nEnter the Product ID of the perticular product you want to buy..")
    if int(ui)>100 and int(ui)<151:
        print('YOUR PRODUCT DETAILS ARE....')
        print("\nType of the product: ", txt1[ui]["Type"])
        print("Colour of the product: ", txt1[ui]["Colour"])
        print("Size of the product: ", txt1[ui]["Size"])
        print("Price of the product: ", txt1[ui]["Price"])

# BILLING part

    apv = int(input("\nType '1' to proceed the Payment... "))
    if apv == 1:
        name = input("\nPlease Enter the Name...")
        quan = int(input("Please Enter the Quantity..."))
        mob = input("Please Enter the Mobile No...")
        adde = input("Please Enter the Address...")
        print("\n")
        print('********YOUR BILL********')
        print('Your Name: ', name)
        print('Your Mobile No.: ', mob)
        print('Your Address: ', adde)
        print("Time of Transection: ", time.ctime())
        print('Type of the product: ', txt1[ui]["Type"])
        print("Price of the product: ", txt1[ui]["Price"]*quan)
        print("Quantity of product: ", quan)
        txt1[ui]['Quantity']= txt1[ui]['Quantity']-quan
        print('*************************')
    
# Updation part of SALES.json

        pro_type = txt1[ui]["Type"]
        pro_price = txt1[ui]["Price"]*quan
        info = {'Customer name': name, 
                'Phone no.': mob, 
                'Address': adde, 
                "Product Type":  pro_type,
                "Product Price": str(pro_price), 
                "Product Quantity ": str(quan)}  
        info2 = json.dumps(info)
        sale = open('Sales.json', 'a')
     
        sale.write(info2 + '\n')
        sale.close()
     
#     Calling update function
        update()
    
    
    else:
        print("**** THANKS FOR VISITING ****")

```

```python
# IN this func all the work is for PRODUCER that he wants to change or update something in the Inventory
def producer():
    print('\nDo you want to UPDATE, CREATE or LOOK at the INVENTORY ')
    print("1: UPDATE")
    print('2: CREATE')
    print("3: LOOK")
    cat = input('\n Enter your choise: ')
    if cat == "1" or cat.lower() == "update":
        ui = input("\nEnter the Product ID of the perticular product you want to Update..")
        if ui > '100' or ui < '151':
            print("Product Information :", txt1[ui])
            print("\nDo you want to ADD the quantity or Change the Price")
            print("1: ADD QUANTITY")
            print('2: CHANGE PRICE')
            chose = input('\n Enter your choise: ')
            if chose == "1" or chose.lower() == "add" or chose.lower()=="add quantity" :
                quan = int(input("Please Enter the Quantity you want to Add..."))
                txt1[ui]['Quantity']= txt1[ui]['Quantity']+quan
                print("Updated Info: "txt1[ui])
                update()
                
            elif chose == "2" or chose.lower() == "change" or chose.lower() == "change price":
                chang = int(input('Enter the price you want to Change...'))
                txt1[ui]["Price"] = chang
                print("Updated Info: "txt1[ui])
                update()
    
        
    elif cat == "2" or cat.lower() == "create":
        prod_id = input("\nEnter the Product ID of the perticular product you want to Create \n It should be greater the '150'..")
        prod_type = input("Enter the product Type: ")
        prod_size = input("Enter the product Size: ")
        prod_color = input("Enter the product Colour: ")
        prod_style = input("Enter the product Style: ")
        prod_price = int(input("Enter the product Price: "))
        prod_quan = int(input("Enter the product quan: "))
        txt2 = {prod_id : {"Type": prod_type, "Size" : prod_size,  "Colour" : prod_color,  "Style" : prod_style , "Price" : prod_price , "Quantity" : prod_quan }}
        ji = json.dumps(txt2)
        fi = open("Inventory.json", "a")
        fi.write(ji + '\n')
        fi.close()
        
    elif cat =='3' or cat.lower()=='look':
        print(txt1)
        
    else:
        print("***** INVALID SELECTION *****")
```

```python
# In this cell we are asking USER that is he a CUSTOMER or PRODUCER and after the selection further operation takes place
print("*** ARE YOU A CUSTOMER OR PRODUCER ***")
identify = int(input("Press '1' for CUSTOMER and Press '0' for PRODUCER.."))
if identify ==1:
    customer()
elif identify ==0:
    print('\nPassword is 1234')
    pas = int(input("Enter the password: "))
    if pas == 1234:
        producer()
    else:
        print("** PASSWORD INCORRECT **")
else:
    print("***** INVALID SELECTION *****")
    
```
